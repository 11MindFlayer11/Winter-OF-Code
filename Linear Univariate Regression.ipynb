```python
#Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
```

```python
#Defining the training function
def train(iterations, learning_rate, X, Y):
  #iterations is the number of times the training is to be done
  #learning rate is the step size
  #X is an array containing the training inputs
  #Y is an array containing true value of training output
  W=1
  b=1

  m=len(X) #defining the learning dataset size
  
   #Defining our model function

    #defining the cost or error function, that is to be minimised

  #defining the gradient functions
  
  iter=[]
  c=[]

  #Applying the gradient descent
  for i in range(iterations):
    Y_pred=X*W+b     #Y_pred is an array with predicted values (size = number of examples)
    gradw=(np.dot((Y_pred-Y),X))/m    #grad w is an array with gradient of the feature
    W=W-learning_rate*gradw
    gradb=(np.sum(Y_pred-Y))/m
    b=b-learning_rate*gradb
    cost=(np.sum((Y_pred-Y)**2))/(2*m)
    iter.append(i)
    c.append(cost)


    if i%5000==0:
      print(cost)
  
  return W,b,iter,c



```

```python
import pandas as pd

df=pd.read_csv('linear_train.csv')
x=df.iloc[:,1]     #assk how to take transpose
X=x.to_numpy()

y=df.iloc[:, 21]
Y=y.to_numpy()

W,b,iter,c=train(500,0.0000242,X,Y)
Y_final=X*W+b
```
